```python
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd

class ReviewSpider(scrapy.Spider):
    name = "Chewy_reviews"
    start_urls = ['https://www.chewy.com/blue-buffalo-life-protection-formula/dp/32041']

    def parse(self, response):
        for review in response.css('div._15b__qxwTXllIZHSTnfiFF p::text').getall():
            yield {
                  #'review_title':  ,
                 'review_content':  review}
            
       # more_reviews= response.css('a.kib-pagination-new-item--next').attrib['href']
       # if next_page is not None:
       #     yield response.follow(next_page, callback=self.parse)

process = CrawlerProcess(settings={
    "FEEDS": {
        "reviews.csv": {"format": "csv"},
    },
})

process.crawl(ReviewSpider)
process.start()
    
```
